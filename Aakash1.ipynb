```python
import cv2
import os

def crop_to_4_3(img):
    h, w = img.shape[:2]
    target_ratio = 4 / 3
    current_ratio = w / h

    if abs(current_ratio - target_ratio) < 1e-3:
        return img  # already 4:3

    if current_ratio > target_ratio:
        # too wide → crop width
        new_w = int(h * target_ratio)
        x_start = (w - new_w) // 2
        cropped = img[:, x_start:x_start + new_w]
    else:
        # too tall → crop height
        new_h = int(w / target_ratio)
        y_start = (h - new_h) // 2
        cropped = img[y_start:y_start + new_h, :]

    return cropped

def process_image(image_path, output_path, max_width=800, max_height=600):
    img = cv2.imread(image_path)
    if img is None:
        print(f"❌ Could not read {image_path}")
        return

    # Step (a): Crop to 4:3
    img = crop_to_4_3(img)
    h, w = img.shape[:2]

    # Step (b, c): Only downscale if larger
    if w > max_width or h > max_height:
        resized = cv2.resize(img, (max_width, max_height), interpolation=cv2.INTER_AREA)
        print(f"✅ Downscaled {os.path.basename(image_path)} → 800×600")
    else:
        resized = img
        print(f"↔️ Skipped (already smaller): {os.path.basename(image_path)} ({w}×{h})")

    cv2.imwrite(output_path, resized)

def batch_process(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)
    supported_exts = ('.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff')

    for filename in os.listdir(input_folder):
        if filename.lower().endswith(supported_exts):
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)
            process_image(input_path, output_path)

    print("\n🎉 All images processed!")

# === Run ===
input_folder = r"C:\Users\Dell\Desktop\IIT\ds203\PROJECT\project_images\images"
output_folder = r"C:\Users\Dell\Desktop\IIT\ds203\PROJECT\project_images\images_resized"

batch_process(input_folder, output_folder)
```
